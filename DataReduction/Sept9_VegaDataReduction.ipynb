```python
image=fits.open('../DatosRaw/Sept9_2024/20240910-000702-Vega-10s-1.fit')
image_data = image[0].data
image.close()
image_array = np.array(image_data)
plt.imshow(image_array, cmap='gray')
plt.colorbar()
plt.show()
```

```python
yvals = np.argmax(image_array, axis=0)
xvals = np.arange(900)
```

```python
plt.plot(xvals,yvals,'x')
plt.ylabel('Argmax trace data')
plt.xlabel('X position')
plt.gca().invert_yaxis()
```

```python
bad_pixels = (yvals < 320) 
plt.plot(xvals, yvals, 'x')
plt.plot(xvals[bad_pixels], yvals[bad_pixels], 'rx')
plt.ylabel("Argmax trace data")
plt.xlabel("X position")
plt.gca().invert_yaxis()
```

```python
plt.plot(xvals[~bad_pixels], yvals[~bad_pixels], 'x')
plt.ylabel("Argmax trace data")
plt.xlabel("X position")
plt.gca().invert_yaxis()
```
